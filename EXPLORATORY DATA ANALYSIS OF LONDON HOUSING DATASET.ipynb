### (A) Convert the Datatype of "Date" column to Date-Time format

```python
data.dtypes                                #BEFORE CONVERSION, THE DATETYPE OF DATE CAN BE SEEN AS OBJECT
```

```python
data.date=pd.to_datetime(data.date)      #CONVERSION
```

```python
data.dtypes                                 #AFTER CONVERSION, THE DATATYPE OF DATE CAN BE SEEN AS DATE TIME
```

### (B.1)  Add a new column "Year" in the dataframe, which contains years only

```python
data["year"]=data.date.dt.year
```

```python
data
```

### (B.2)  Add a new column "month" in the dataframe, which contains month only

```python
data.insert(1, "month", data.date.dt.month)
```

```python
data
```

### (C) Remove the columns "year" and "month" from the dataframe

```python
data.drop(["month","year"],axis=1,inplace=True)
```

```python
data
```

### (D) Show all the records where "no_of_crimes" is 0 and how many of such records are there?

### (D) What is the "average_price" per year in England?

```python
data["year"]=data.date.dt.year
```

```python
data
```

```python
data.groupby("year").average_price.mean()
```

#### As the years go by, there seem to be an increase in the average house price in England

### (E) What is the maximum and minimum number of crimes recorded per area?

```python
data.groupby("area").no_of_crimes.max().sort_values(ascending=False) #MAXINUM NUMBER OF CRIMES RECORDED PER AREA
```

```python
data.groupby("area").no_of_crimes.min().sort_values(ascending=True)  #MINIMUM NUMBER OF CRIMES PER AREA
```

### (F) Show relationships between variables
